```python
url = 'https://data.cityofnewyork.us/resource/nwxe-4ae8.json'
trees = pd.read_json(url)
trees.head(10)
```

```python
trees = trees[['borocode', 'boroname', 'tree_id', 'steward', 'latitude', 'longitude', 'health', 'spc_common']]
trees['steward'] = trees['steward'].map({'None': 'None', '1or2': 'Some', '3or4': 'Many', '4orMore': 'Most', np.nan: 'None'})
```

```python
trees['steward'].unique()
```

```python
trees.head()
```

```python
with open('trees.csv', 'w') as write:
    write.write(trees.to_csv())
```

```python
spc = trees.groupby(['spc_common', 'health'])[['tree_id']].count().reset_index()
spc.head(10)
```

```python
steward = trees.groupby(['steward', 'health'])['tree_id'].count().reset_index()
steward['steward'] = pd.Categorical(steward['steward'], ['None', 'Some', 'Many', 'Most'])
steward = steward.sort_values(['steward', 'health'])
```
